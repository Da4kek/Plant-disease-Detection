```python
import pandas as pd 
import matplotlib.pyplot as plt 
from matplotlib.image import imread
import numpy as np
import os
%matplotlib inline
```

```python
data_dir = "dataset/dataset/"
print(os.listdir(data_dir))
train_dir = os.path.join(data_dir, 'train/')
test_dir = os.path.join(data_dir, 'test/')
print(train_dir)
print(os.listdir(train_dir))
print(os.listdir(test_dir))
```

```python
test_categories = os.listdir(test_dir)
train_categories = os.listdir(train_dir)

print("Categories in test directory:", test_categories)
print("Categories in train directory:", train_categories)
```

```python
for category in train_categories:
    category_dir = os.path.join(train_dir, category)
    num_images = len(os.listdir(category_dir))
    print(f"Number of images in {category}: {num_images}")

for category in test_categories:
    category_dir = os.path.join(test_dir, category)
    num_images = len(os.listdir(category_dir))
    print(f"Number of images in {category}: {num_images}")
```

```python
black_spot = []
citrus_canker = []

for img in os.listdir(os.path.join(train_dir,'black-spot')):
    black_spot.append(imread(os.path.join(train_dir,'black-spot',img)))

for img in os.listdir(os.path.join(train_dir,'citrus-canker')):
    citrus_canker.append(imread(os.path.join(train_dir,'citrus-canker',img)))
```

```python
print(black_spot[0].shape)
print(citrus_canker[0].shape)
```

```python
print(len(black_spot))
print(len(citrus_canker))
```

```python
fig, axes = plt.subplots(1, 2)

axes[0].imshow(black_spot[0])
axes[0].set_title('Black Spot')

axes[1].imshow(citrus_canker[0])
axes[1].set_title('Citrus Canker')

plt.show()
```

```python
import random 
fig, axes = plt.subplots(1, 2)
axes[0].imshow(black_spot[random.randint(0, len(black_spot))])
axes[0].set_title('Black Spot')

axes[1].imshow(citrus_canker[random.randint(0, len(citrus_canker))])
axes[1].set_title('Citrus Canker');
```

```python
img_width = []
img_height = []

for i in range(len(black_spot)):
    img_width.append(black_spot[i].shape[0])
    img_height.append(black_spot[i].shape[1])

print(np.average(img_width))
print(np.average(img_height))
```
