```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.image import imread
%matplotlib inline
```

```python
import os
dataset_dir = "dataset/dataset/"
print(os.listdir(dataset_dir))
```

```python
from cnn_architecture import CNN
train_dir = "dataset/dataset/train"
test_dir = "dataset/dataset/test"
model = CNN(train_dir=train_dir, test_dir=test_dir)
```

```python
model_hist = model.train(10)
```

```python
loss = model_hist.history['loss']
accuracy = model_hist.history['accuracy']

epochs = range(1, len(loss) + 1)

plt.plot(epochs, loss, 'b', label='Loss')
plt.plot(epochs, accuracy, 'r', label='Accuracy')
plt.title('Training Loss and Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
model.save_model("models/cnn_model.h5")
```
